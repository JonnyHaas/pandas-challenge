The Analysis of the PyCitySchools reveals the requested information about the schools and draws two conclusions.  The two conclusions are listed at the bottom of this report.

The Analysis:

Overall District Summary:  15 schools with 39,170 students with overall passing % of 65.2%.

Individual Schools Summary:  In general District schools are much larger than Charter schools with a much lower passing percentage.

Top 5 schools by overall passing percentage (reading and math):  Top 5 schools are all Charter with over 90% passing percentage.

Bottom 5 schools by overall passing percentage (reading and math): Bottom 5 schools are all District schools with barely over a 50% passing percentage.

Math Scores by Grade:  Scores stay pretty consistent per grades stay for every school.

Reading Scores by Grade:  Scores stay pretty consistent per grades stay for every school.

Scores by Spending Range:  More spending per student does not equate to higher scores.  In fact the inverse.

Scores by school size:  Small and Medium schools do much better than large schools.

Spending Summary:  Charter schools scores are much higher than District schools with lower spending cost per student.

School Type Summary:  Charter schools have a much higher passing rate than District schools.

Conclusion:
1.	Charter Schools have much higher passing rates than District schools on generally lower budgets per student.  
2.	Charter Schools are much smaller than District Schools which would appear to correlate to higher passing rates and lower budgets per student.



```python
import pandas as pd

# File paths
schools_file_path = 'Resources/Schools_complete.csv'
students_file_path = 'Resources/Students_complete.csv'

# Read in the CSV files
schools_df = pd.read_csv(schools_file_path)
students_df = pd.read_csv(students_file_path)

# Merge the dataframes on 'school_name'
school_data_complete = pd.merge(students_df, schools_df, how='left', on=['school_name', 'school_name'])

# Total number of unique schools
total_schools = schools_df['School ID'].nunique()

# Total number of students
total_students = students_df['Student ID'].nunique()

# Total budget
total_budget = schools_df['budget'].sum()

# Average math score
average_math_score = students_df['math_score'].mean()

# Average reading score
average_reading_score = students_df['reading_score'].mean()

# Percentage passing math
passing_math_count = students_df[students_df['math_score'] >= 70].shape[0]
percent_passing_math = (passing_math_count / total_students) * 100

# Percentage passing reading
passing_reading_count = students_df[students_df['reading_score'] >= 70].shape[0]
percent_passing_reading = (passing_reading_count / total_students) * 100

# Overall passing rate (students who passed both math and reading)
overall_passing_count = students_df[(students_df['math_score'] >= 70) & (students_df['reading_score'] >= 70)].shape[0]
percent_overall_passing = (overall_passing_count / total_students) * 100

# Create a DataFrame for the district summary
district_summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "Percent Passing Math": [percent_passing_math],
    "Percent Passing Reading": [percent_passing_reading],
    "Percent Overall Passing": [percent_overall_passing]
})

# Display the DataFrame
district_summary_df
```

```python
import pandas as pd

# Group data by school
grouped_school_data = school_data_complete.groupby('school_name')

# School type (using the first entry for each school as all entries are the same for each school)
school_type = grouped_school_data['type'].first()

# Total students per school
total_students_per_school = grouped_school_data['Student ID'].count()

# Total school budget per school
total_school_budget = grouped_school_data['budget'].mean()  # Assuming budget is constant per school

# Per student budget
per_student_budget = total_school_budget / total_students_per_school

# Average math score per school
average_math_score = grouped_school_data['math_score'].mean()

# Average reading score per school
average_reading_score = grouped_school_data['reading_score'].mean()

# Calculate percentages for math, reading, and overall passing
passing_math = school_data_complete[school_data_complete['math_score'] >= 70]
passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70]
overall_passing = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]

# Percent passing math
percent_passing_math = passing_math.groupby('school_name')['Student ID'].count() / total_students_per_school * 100

# Percent passing reading
percent_passing_reading = passing_reading.groupby('school_name')['Student ID'].count() / total_students_per_school * 100

# Overall percent passing
percent_overall_passing = overall_passing.groupby('school_name')['Student ID'].count() / total_students_per_school * 100

# Create a DataFrame to hold the results
school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_students_per_school,
    "Total School Budget": total_school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing": percent_overall_passing
})

# Display the DataFrame
school_summary_df
```

```python
# Sort the school summary DataFrame by '% Overall Passing' in descending order
top_schools = school_summary_df.sort_values("% Overall Passing", ascending=False)

# Display the top 5 rows
top_schools.head()
```
