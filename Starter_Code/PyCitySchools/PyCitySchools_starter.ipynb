```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
# district_summary = 

# Formatting
# district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
# district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
# district_summary

import pandas as pd

# Assuming school_data_complete is your merged DataFrame from the previous steps

# Group data by school
grouped_school_data = school_data_complete.groupby('school_name')

# Total students per school
total_students_per_school = grouped_school_data['student_name'].count()

# Total school budget per school
total_school_budget = grouped_school_data['budget'].mean()  # Assuming budget is constant per school

# Per student budget
per_student_budget = total_school_budget / total_students_per_school

# Average math score per school
average_math_score = grouped_school_data['math_score'].mean()

# Average reading score per school
average_reading_score = grouped_school_data['reading_score'].mean()

# Calculate percentages
passing_math = school_data_complete[school_data_complete['math_score'] >= 70]
passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70]
passing_both = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]

# Percentage passing math per school
percent_passing_math = passing_math.groupby('school_name')['student_name'].count() / total_students_per_school * 100

# Percentage passing reading per school
percent_passing_reading = passing_reading.groupby('school_name')['student_name'].count() / total_students_per_school * 100

# Overall passing rate (average of math and reading passing rates)
percent_overall_passing = passing_both.groupby('school_name')['student_name'].count() / total_students_per_school * 100

# Create a DataFrame to hold the results
school_summary_df = pd.DataFrame({
    "Total Students": total_students_per_school,
    "Total School Budget": total_school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing": percent_overall_passing
})

# Formatting the DataFrame for better readability
formatted_school_summary_df = school_summary_df.style.format({
    "Total School Budget": "${:,.2f}",
    "Per Student Budget": "${:,.2f}",
    "Average Math Score": "{:.1f}",
    "Average Reading Score": "{:.1f}",
    "% Passing Math": "{:.2f}%",
    "% Passing Reading": "{:.2f}%",
    "% Overall Passing": "{:.2f}%"
})

# Display the formatted DataFrame
formatted_school_summary_df
```
